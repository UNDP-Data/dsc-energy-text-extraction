```python
import pandas as pd
import numpy as np
import json
import os
```

```python
with open('document_compilation_json/documents.json', 'r') as json_file:
    data = json.load(json_file)
df = pd.DataFrame(data)
df
```

```python
def remove_symbol(row):
    return row['Content'].replace('\uf0a7', ';')
df['Output_content'] = df.apply(lambda row: remove_symbol(row), axis=1)
df['Year'] = df['Code'].str.split('-').str.get(2)
```

```python
import json
import os
import math

output_directory = 'txt_files_raw'
os.makedirs(output_directory, exist_ok=True)

for index, row in df.iterrows():
    name = row['Code']
    filename = os.path.join(output_directory, f'{name}.txt')
    with open(filename, 'w') as txt_file:
        txt_file.write(f"File Name: {name}\n"
                       f"Year: {row['Year']}\n"
                       f"Country Name: {row['Country Name']}\n"
                       f"Country Code: {row['Country Code']}\n"
                       f"Category: {row['Category']}\n"
                       f"Document Title: {row['Document Title']}\n"
                       f"Exists?: {row['Exists?']}\n"
                       f"Publication Date: {row['Publication Date']}\n"
                       f"Start Year: {row['Start Year']}\n"
                       f"End Year: {row['End Year']}\n"
                       f"Language: {row['Language']}\n"
                       f"------------------------------------------------\n\n")
        
        txt_file.write(row['Output_content'])

print("Text files created and distributed successfully.")
```
